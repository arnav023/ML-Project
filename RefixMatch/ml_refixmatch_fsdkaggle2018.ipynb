```python
import os
import torch
import torchaudio
import torchaudio.transforms as T
import random
import numpy as np
import librosa
import matplotlib.pyplot as plt

from torch import nn, optim
from torchvision import models, transforms
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import pandas as pd

#config
SEED = 42
SAMPLE_RATE = 16000
NUM_CLASSES = 41
BATCH_SIZE = 64
NUM_EPOCHS = 5
THRESHOLD = 0.95
LABELED_SAMPLES_PER_CLASS = [5, 20, 30, 40]
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.manual_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

audio_zip = 'FSDKaggle2018.audio_train.zip'
meta_zip = 'FSDKaggle2018.meta.zip'
audio_dir = 'FSDKaggle2018.audio_train'
meta_dir = 'FSDKaggle2018.meta'

if not os.path.exists(audio_zip) or not os.path.exists(meta_zip):
    print("Downloading the dataset...")
    !wget https://zenodo.org/record/2552860/files/FSDKaggle2018.audio_train.zip?download=1 -O {audio_zip}
    !wget https://zenodo.org/record/2552860/files/FSDKaggle2018.meta.zip?download=1 -O {meta_zip}
else:
    print("Dataset zip files already exist. Skipping download.")

if not os.path.exists(audio_dir) or not os.path.exists(meta_dir):
    print("Unzipping the dataset...")
    !unzip -q {audio_zip}
    !unzip -q {meta_zip}
else:
    print("Dataset already unzipped. Skipping extraction.")

meta_df = pd.read_csv('FSDKaggle2018.meta/train_post_competition.csv')
label_encoder = LabelEncoder()
meta_df = meta_df[meta_df['manually_verified'] == 1].reset_index(drop=True)
meta_df['encoded'] = label_encoder.fit_transform(meta_df['label'])  #updated to label

def augment_weak(mel):
    return mel + 0.01 * torch.randn_like(mel)

def augment_strong(mel):
    mel = T.FrequencyMasking(freq_mask_param=15)(mel)
    mel = T.TimeMasking(time_mask_param=35)(mel)
    return mel

mel_transform = T.MelSpectrogram(
    sample_rate=SAMPLE_RATE,
    n_fft=1024,
    hop_length=512,
    n_mels=128
)

class FSDDataset(Dataset):
    def __init__(self, df, audio_dir, labeled=True, transform=None):
        self.df = df
        self.audio_dir = audio_dir
        self.labeled = labeled
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        filepath = os.path.join(self.audio_dir, row['fname'])
        waveform, sr = torchaudio.load(filepath)
        waveform = torchaudio.transforms.Resample(orig_freq=sr, new_freq=SAMPLE_RATE)(waveform)
        waveform = waveform.mean(dim=0, keepdim=True)
        if waveform.shape[1] < SAMPLE_RATE:
            waveform = torch.nn.functional.pad(waveform, (0, SAMPLE_RATE - waveform.shape[1]))
        else:
            waveform = waveform[:, :SAMPLE_RATE]

        mel = mel_transform(waveform)
        mel = torch.log1p(mel)
        mel = (mel - mel.mean()) / mel.std()
        mel = mel.expand(3, -1, -1)

        if self.transform:
            mel = self.transform(mel)

        label = row['encoded'] if self.labeled else -1
        return mel, label

def get_labeled_unlabeled(df, samples_per_class):
    labeled_indices = []
    for label in df['encoded'].unique():
        label_df = df[df['encoded'] == label]
        if len(label_df) >= samples_per_class:
            labeled_indices.extend(label_df.sample(samples_per_class, random_state=SEED).index.tolist())
    labeled_df = df.loc[labeled_indices].reset_index(drop=True)
    unlabeled_df = df.drop(index=labeled_indices).reset_index(drop=True)
    return labeled_df, unlabeled_df

def get_model():
    model = models.mobilenet_v2(pretrained=True)
    model.classifier[1] = nn.Linear(model.classifier[1].in_features, NUM_CLASSES)
    return model.to(DEVICE)

def train_refixmatch(labeled_loader, unlabeled_loader, val_loader, label_setting):
    model = get_model()
    optimizer = optim.Adam(model.parameters(), lr=1e-4)
    criterion = nn.CrossEntropyLoss()
    best_f1 = 0
    train_losses, val_losses = [], []

    for epoch in range(NUM_EPOCHS):
        model.train()
        total_loss = 0

        for (x_lb, y_lb), (x_ulb, _) in zip(labeled_loader, unlabeled_loader):
            x_lb, y_lb = x_lb.to(DEVICE), y_lb.to(DEVICE)
            x_ulb_weak = augment_weak(x_ulb).to(DEVICE)
            x_ulb_strong = augment_strong(x_ulb).to(DEVICE)

            with torch.no_grad():
                model.eval()
                preds = torch.softmax(model(x_ulb_weak), dim=1)
                max_probs, pseudo_labels = torch.max(preds, dim=1)
                mask = max_probs.ge(THRESHOLD).float()

            model.train()
            out_lb = model(x_lb)
            out_ulb = model(x_ulb_strong)

            loss_lb = criterion(out_lb, y_lb)
            loss_ulb = (criterion(out_ulb, pseudo_labels) * mask).mean()
            loss = loss_lb + loss_ulb

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            total_loss += loss.item()

        train_losses.append(total_loss / len(labeled_loader))

        model.eval()
        val_loss = 0
        all_preds, all_labels = [], []

        with torch.no_grad():
            for x, y in val_loader:
                x, y = x.to(DEVICE), y.to(DEVICE)
                out = model(x)
                val_loss += criterion(out, y).item()
                preds = out.argmax(1)
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(y.cpu().numpy())

        val_losses.append(val_loss / len(val_loader))

        acc = accuracy_score(all_labels, all_preds)
        prec = precision_score(all_labels, all_preds, average='macro', zero_division=0)
        rec = recall_score(all_labels, all_preds, average='macro', zero_division=0)
        f1 = f1_score(all_labels, all_preds, average='macro', zero_division=0)

        print(f"[Label={label_setting}] Epoch {epoch+1}: Loss={total_loss:.4f} | Acc={acc:.4f} | Prec={prec:.4f} | Rec={rec:.4f} | F1={f1:.4f}")

        if f1 > best_f1:
            best_f1 = f1
            torch.save(model.state_dict(), f"best_model_mobilenet_refixmatch_{label_setting}.pt")

    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Val Loss')
    plt.title(f"Loss Curves - {label_setting} labeled/class")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid(True)
    plt.show()

for n_labeled in LABELED_SAMPLES_PER_CLASS:
    labeled_df, unlabeled_df = get_labeled_unlabeled(meta_df, n_labeled)
    val_df = labeled_df.sample(frac=0.2, random_state=SEED)
    train_df = labeled_df.drop(val_df.index)

    labeled_ds = FSDDataset(train_df, 'FSDKaggle2018.audio_train', labeled=True)
    unlabeled_ds = FSDDataset(unlabeled_df, 'FSDKaggle2018.audio_train', labeled=False)
    val_ds = FSDDataset(val_df, 'FSDKaggle2018.audio_train', labeled=True)

    labeled_loader = DataLoader(labeled_ds, batch_size=BATCH_SIZE, shuffle=True)
    unlabeled_loader = DataLoader(unlabeled_ds, batch_size=BATCH_SIZE, shuffle=True)
    val_loader = DataLoader(val_ds, batch_size=BATCH_SIZE)

    train_refixmatch(labeled_loader, unlabeled_loader, val_loader, label_setting=n_labeled)
```

```python

def evaluate_model(model, test_loader):
    model.eval()
    all_preds, all_labels = [], []
    criterion = nn.CrossEntropyLoss()

    #running on test set
    with torch.no_grad():
        test_loss = 0
        for x, y in test_loader:
            x, y = x.to(DEVICE), y.to(DEVICE)
            out = model(x)
            test_loss += criterion(out, y).item()
            preds = out.argmax(1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(y.cpu().numpy())

    acc = accuracy_score(all_labels, all_preds)
    prec = precision_score(all_labels, all_preds, average='macro', zero_division=0)
    rec = recall_score(all_labels, all_preds, average='macro', zero_division=0)
    f1 = f1_score(all_labels, all_preds, average='macro', zero_division=0)

    from sklearn.metrics import roc_auc_score
    try:
        auc_roc = roc_auc_score(all_labels, np.array(all_preds), multi_class='ovr')
    except ValueError:
        auc_roc = 'undefined'

    print(f"Test Loss: {test_loss / len(test_loader):.4f}")
    print(f"Accuracy: {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall: {rec:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print(f"AUC-ROC: {auc_roc}")

best_model_path = "best_model_mobilenet_refixmatch_40.pt"
model = get_model()
model.load_state_dict(torch.load(best_model_path))
model.to(DEVICE)

test_df = meta_df.sample(frac=0.2, random_state=SEED)
test_ds = FSDDataset(test_df, 'FSDKaggle2018.audio_train', labeled=True)
test_loader = DataLoader(test_ds, batch_size=BATCH_SIZE)

evaluate_model(model, test_loader)
```
