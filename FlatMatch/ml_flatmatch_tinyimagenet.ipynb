```python
import os
import random
import zipfile
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import requests
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, Subset
from torchvision import models, transforms, datasets
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, average_precision_score

#setting up
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
os.makedirs("saved_models", exist_ok=True)
os.makedirs("loss_curves", exist_ok=True)

def set_seed(seed=42):
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    random.seed(seed)

set_seed()


def download_and_extract_tiny_imagenet(dest="/content"):
    url = "http://cs231n.stanford.edu/tiny-imagenet-200.zip"
    zip_path = os.path.join(dest, "tiny-imagenet-200.zip")
    extracted_path = os.path.join(dest, "tiny-imagenet-200")

    if not os.path.exists(extracted_path):
        print("Downloading Tiny ImageNet...")
        response = requests.get(url, stream=True)
        with open(zip_path, "wb") as f:
            for chunk in tqdm(response.iter_content(1024)):
                f.write(chunk)

        print("Extracting...")
        with zipfile.ZipFile(zip_path, "r") as zip_ref:
            zip_ref.extractall(dest)

        os.remove(zip_path)
    else:
        print("Tiny ImageNet already downloaded.")

    return extracted_path

root_dir = download_and_extract_tiny_imagenet()

#augmentations
weak_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(64, padding=4),
    transforms.ToTensor(),
])

strong_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(64, padding=4),
    transforms.ColorJitter(0.4, 0.4, 0.4, 0.1),
    transforms.RandomGrayscale(p=0.2),
    transforms.ToTensor(),
])

val_transform = transforms.ToTensor()

train_dataset = datasets.ImageFolder(root=os.path.join(root_dir, "train"), transform=weak_transform)

class TinyImageNetValDataset(torch.utils.data.Dataset):
    def __init__(self, root_dir, transform=None, class_to_idx=None):
        self.root_dir = os.path.join(root_dir, "val")
        self.transform = transform
        self.img_dir = os.path.join(self.root_dir, "images")
        self.image_labels = []
        self.class_to_idx = class_to_idx  # use train mapping

        with open(os.path.join(self.root_dir, "val_annotations.txt")) as f:
            for line in f:
                img_file, class_id = line.strip().split("\t")[:2]
                label = self.class_to_idx[class_id]  # map using train class_to_idx
                self.image_labels.append((img_file, label))

    def __len__(self):
        return len(self.image_labels)

    def __getitem__(self, idx):
        img_file, label = self.image_labels[idx]
        img_path = os.path.join(self.img_dir, img_file)
        image = Image.open(img_path).convert("RGB")
        if self.transform:
            image = self.transform(image)
        return image, label
val_dataset = TinyImageNetValDataset(root_dir, transform=val_transform, class_to_idx=train_dataset.class_to_idx)


def split_labeled_unlabeled(dataset, num_labeled):
    labels = np.array([label for _, label in dataset])
    labeled_idx, unlabeled_idx = [], []
    for c in np.unique(labels):
        idx = np.where(labels == c)[0]
        np.random.shuffle(idx)
        labeled_count = num_labeled // 200
        labeled_idx.extend(idx[:labeled_count])
        unlabeled_idx.extend(idx[labeled_count:])
    return Subset(dataset, labeled_idx), Subset(dataset, unlabeled_idx)


class FlatMatchTrainer:
    def __init__(self, model, labeled_loader, unlabeled_loader, val_loader,
                 num_classes=200, threshold=0.95, lambda_u=1.0, num_epochs=10):
        self.model = model.to(device)
        self.labeled_loader = labeled_loader
        self.unlabeled_loader = unlabeled_loader
        self.val_loader = val_loader
        self.criterion = nn.CrossEntropyLoss()
        self.optimizer = optim.Adam(self.model.parameters(), lr=5e-4)
        self.threshold = threshold
        self.lambda_u = lambda_u
        self.num_epochs = num_epochs

    def train(self):
        train_losses, val_losses = [], []
        best_f1 = -1

        for epoch in range(self.num_epochs):
            self.model.train()
            total_loss = 0
            unlabeled_iter = iter(self.unlabeled_loader)

            for x_l, y_l in self.labeled_loader:
                try:
                    x_u_w, _ = next(unlabeled_iter)
                except StopIteration:
                    unlabeled_iter = iter(self.unlabeled_loader)
                    x_u_w, _ = next(unlabeled_iter)

                x_l, y_l = x_l.to(device), y_l.to(device)
                x_u_w = x_u_w.to(device)
                x_u_s = torch.stack([strong_transform(Image.fromarray((img.cpu().numpy().transpose(1, 2, 0) * 255).astype(np.uint8))) for img in x_u_w])

                #labeled loss
                logits_l = self.model(x_l)
                loss_l = self.criterion(logits_l, y_l)

                #unlabeled loss
                with torch.no_grad():
                    logits_uw = self.model(x_u_w)
                    probs_uw = torch.softmax(logits_uw, dim=1)
                    avg_probs = probs_uw.mean(dim=0)
                    probs_uw = probs_uw / avg_probs
                    probs_uw = probs_uw / probs_uw.sum(dim=1, keepdim=True)

                    max_probs, targets_u = torch.max(probs_uw, dim=1)
                    mask = max_probs.ge(self.threshold).float()

                logits_us = self.model(x_u_s.to(device))
                loss_u = (F.cross_entropy(logits_us, targets_u.to(device), reduction='none') * mask).mean()

                loss = loss_l + self.lambda_u * loss_u

                self.optimizer.zero_grad()
                loss.backward()
                self.optimizer.step()
                total_loss += loss.item()

            train_losses.append(total_loss)
            val_loss, f1 = self.evaluate()
            val_losses.append(val_loss)

            print(f"Epoch {epoch+1}: Train Loss={total_loss:.4f}, Val Loss={val_loss:.4f}, F1={f1:.4f}")

            if f1 > best_f1:
                best_f1 = f1
                torch.save(self.model.state_dict(), "saved_models/best_flatmatch_model.pth")

        return train_losses, val_losses

    def evaluate(self):
        self.model.eval()
        y_true, y_pred, y_prob = [], [], []
        total_loss = 0

        with torch.no_grad():
            for x, y in self.val_loader:
                x, y = x.to(device), y.to(device)
                outputs = self.model(x)
                loss = self.criterion(outputs, y)
                total_loss += loss.item()
                probs = torch.softmax(outputs, dim=1)
                preds = torch.argmax(probs, dim=1)
                y_true.extend(y.cpu().numpy())
                y_pred.extend(preds.cpu().numpy())
                y_prob.extend(probs.cpu().numpy())

        acc = accuracy_score(y_true, y_pred)
        prec = precision_score(y_true, y_pred, average='macro', zero_division=0)
        rec = recall_score(y_true, y_pred, average='macro', zero_division=0)
        f1 = f1_score(y_true, y_pred, average='macro', zero_division=0)
        try:
            auc_roc = roc_auc_score(y_true, y_prob, multi_class='ovo')
            auc_pr = average_precision_score(y_true, y_prob, average='macro')
        except:
            auc_roc, auc_pr = -1, -1

        print(f"  Accuracy:  {acc:.4f}")
        print(f"  Precision: {prec:.4f}")
        print(f"  Recall:    {rec:.4f}")
        print(f"  F1 Score:  {f1:.4f}")
        print(f"  AUC ROC:   {auc_roc:.4f}")
        print(f"  AUC PRC:   {auc_pr:.4f}")
        return total_loss, f1

label_counts = [10000, 5000, 2000, 400]
batch_size = 64
epochs = 5

for num_labels in label_counts:
    print(f"\nRunning FlatMatch with {num_labels} labeled samples...")

    labeled_subset, unlabeled_subset = split_labeled_unlabeled(train_dataset, num_labels)
    labeled_loader = DataLoader(labeled_subset, batch_size=batch_size, shuffle=True)
    unlabeled_loader = DataLoader(unlabeled_subset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

    model = models.resnet18(pretrained=True)
    model.fc = nn.Linear(model.fc.in_features, 200)

    trainer = FlatMatchTrainer(model, labeled_loader, unlabeled_loader, val_loader,
                               num_classes=200, threshold=0.95, lambda_u=1.0, num_epochs=epochs)

    train_losses, val_losses = trainer.train()

    plt.figure()
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Val Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.title(f"Loss Curves for {num_labels} Labeled Samples")
    plt.savefig(f"loss_curves/loss_curve_{num_labels}_labeled.png")
    plt.show()

    print(f"Finished training for {num_labels} labeled samples.")
```

```python
import os
import torch
from tqdm import tqdm
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms, models
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, average_precision_score
from PIL import Image
import numpy as np

val_transform = transforms.Compose([
    transforms.Resize(64),
    transforms.CenterCrop(64),
    transforms.ToTensor(),
])

def get_class_to_idx(root_dir):
    wnid_file = os.path.join(root_dir, "tiny-imagenet-200", "wnids.txt")
    with open(wnid_file, "r") as f:
        wnids = [line.strip() for line in f.readlines()]
    wnid_to_idx = {wnid: idx for idx, wnid in enumerate(wnids)}
    return wnid_to_idx

class TinyImageNetValDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = os.path.join(root_dir, "tiny-imagenet-200", "val")
        self.transform = transform
        self.class_to_idx = get_class_to_idx(root_dir)

        self.img_labels = []
        with open(os.path.join(self.root_dir, "val_annotations.txt"), "r") as f:
            for line in f.readlines():
                parts = line.strip().split('\t')
                filename, wnid = parts[0], parts[1]

                if wnid in self.class_to_idx:
                    self.img_labels.append((filename, self.class_to_idx[wnid]))

    def __len__(self):
        return len(self.img_labels)

    def __getitem__(self, idx):
        filename, label = self.img_labels[idx]
        img_path = os.path.join(self.root_dir, "images", filename)
        image = Image.open(img_path).convert("RGB")
        if self.transform:
            image = self.transform(image)
        return image, label

val_dataset = TinyImageNetValDataset(root_dir='/content', transform=val_transform)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = models.resnet18(weights=None)
model.fc = torch.nn.Linear(model.fc.in_features, 200)
model.load_state_dict(torch.load("saved_models/best_flatmatch_model.pth"))
model = model.to(device)
model.eval()

def evaluate():
    y_true, y_pred, y_prob = [], [], []

    with torch.no_grad():
        for images, labels in tqdm(val_loader):
            images, labels = images.to(device), labels.to(device)

            outputs = model(images)
            probs = torch.softmax(outputs, dim=1)

            preds = torch.argmax(probs, dim=1)

            y_true.extend(labels.cpu().numpy())
            y_pred.extend(preds.cpu().numpy())
            y_prob.extend(probs.cpu().numpy())

    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, average='macro', zero_division=0)
    rec = recall_score(y_true, y_pred, average='macro', zero_division=0)
    f1 = f1_score(y_true, y_pred, average='macro', zero_division=0)

    try:
        auc_roc = roc_auc_score(y_true, y_prob, multi_class='ovo')
        auc_pr = average_precision_score(y_true, y_prob, average='macro')
    except:
        auc_roc, auc_pr = -1, -1

    print(f"Accuracy:  {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall:    {rec:.4f}")
    print(f"F1 Score:  {f1:.4f}")
    print(f"AUC ROC:   {auc_roc:.4f}")
    print(f"AUC PRC:   {auc_pr:.4f}")

evaluate()
```
