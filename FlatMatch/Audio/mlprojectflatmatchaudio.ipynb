```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
)
import seaborn as sns
import random
from tqdm import tqdm
import librosa
import torchaudio
import torchvision.models as models

# Set seeds for reproducibility
random.seed(42)
np.random.seed(42)
torch.manual_seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(42)

# Create output directory
OUTPUT_DIR = "Flatmatch_AudioData"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Define device
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Paths
DATA_PATH = "/kaggle/input/filtereddata/audio"
TRAIN_PATH = os.path.join(DATA_PATH, "train")
TEST_PATH = os.path.join(DATA_PATH, "test")
VAL_PATH = os.path.join(DATA_PATH, "val")
TRAIN_METADATA = os.path.join(DATA_PATH, "train_metadata.csv")
TEST_METADATA = os.path.join(DATA_PATH, "test_metadata.csv")
VAL_METADATA = os.path.join(DATA_PATH, "val_metadata.csv")

# -----------------------------------------------------
# EMA Implementation
# -----------------------------------------------------
class EMA:
    def __init__(self, model, decay):
        self.model = model
        self.decay = decay
        self.shadow = {}
        self.backup = {}
        self.register()

    def register(self):
        for name, param in self.model.named_parameters():
            if param.requires_grad:
                self.shadow[name] = param.data.clone()

    def update(self):
        for name, param in self.model.named_parameters():
            if param.requires_grad:
                assert name in self.shadow
                new_average = (
                    1.0 - self.decay
                ) * param.data + self.decay * self.shadow[name]
                self.shadow[name] = new_average.clone()

    def apply_shadow(self):
        for name, param in self.model.named_parameters():
            if param.requires_grad:
                assert name in self.shadow
                self.backup[name] = param.data.clone()
                param.data = self.shadow[name]

    def restore(self):
        for name, param in self.model.named_parameters():
            if param.requires_grad:
                assert name in self.backup
                param.data = self.backup[name]
        self.backup = {}


# -----------------------------------------------------
# Audio Feature Extraction and Dataset
# -----------------------------------------------------
def load_audio_file(file_path, sample_rate=22050, duration=5):
    """Load audio file and process it to fixed length."""
    try:
        # Load audio with librosa
        audio, sr = librosa.load(file_path, sr=sample_rate, mono=True)

        # Ensure fixed length
        target_length = int(sample_rate * duration)

        # If audio is too short, pad it
        if len(audio) < target_length:
            audio = np.pad(audio, (0, target_length - len(audio)), mode="constant")

        # If audio is too long, truncate it
        else:
            audio = audio[:target_length]

        return audio
    except Exception as e:
        print(f"Error loading {file_path}: {e}")
        # Return silent audio in case of error
        return np.zeros(int(sample_rate * duration))


def extract_features(audio, sample_rate=22050, n_mels=128, n_fft=2048, hop_length=512):
    """Extract mel spectrogram features from audio."""
    # Compute mel spectrogram
    mel_spec = librosa.feature.melspectrogram(
        y=audio, sr=sample_rate, n_mels=n_mels, n_fft=n_fft, hop_length=hop_length
    )

    # Convert to decibels
    mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)

    # Normalize
    mel_spec_db_norm = (mel_spec_db - mel_spec_db.mean()) / (mel_spec_db.std() + 1e-8)

    return mel_spec_db_norm


class AudioDataset(Dataset):
    def __init__(
        self,
        metadata_path,
        audio_dir,
        transform=None,
        num_samples_per_class=None,
        preprocessing=None,
    ):
        self.metadata = pd.read_csv(metadata_path)
        self.audio_dir = audio_dir
        self.transform = transform
        self.preprocessing = preprocessing

        # Get mapping of class names to indices
        self.class_names = sorted(self.metadata["label"].unique())
        self.class_to_idx = {cls_name: i for i, cls_name in enumerate(self.class_names)}
        self.num_classes = len(self.class_names)

        print(f"Found {self.num_classes} classes: {self.class_names}")

        # Filter by number of samples per class if specified
        if num_samples_per_class:
            # Group by class and select limited samples
            grouped = self.metadata.groupby("label")
            selected_data = []

            for class_name, group in grouped:
                if len(group) > num_samples_per_class:
                    selected_data.append(
                        group.sample(num_samples_per_class, random_state=42)
                    )
                else:
                    selected_data.append(group)

            self.metadata = pd.concat(selected_data).reset_index(drop=True)
            print(
                f"Filtered to {num_samples_per_class} samples per class. New dataset size: {len(self.metadata)}"
            )

        # Verify files exist
        valid_files = []
        for i, row in self.metadata.iterrows():
            file_path = os.path.join(self.audio_dir, row["fname"])
            if os.path.exists(file_path):
                valid_files.append(i)
            else:
                print(f"Warning: File not found - {file_path}")

        self.metadata = self.metadata.iloc[valid_files].reset_index(drop=True)
        print(f"Final dataset size after filtering missing files: {len(self.metadata)}")

    def __len__(self):
        return len(self.metadata)

    def __getitem__(self, idx):
        file_name = self.metadata.iloc[idx]["fname"]
        file_path = os.path.join(self.audio_dir, file_name)
        class_name = self.metadata.iloc[idx]["label"]
        class_idx = self.class_to_idx[class_name]

        # Load and preprocess audio
        audio = load_audio_file(file_path)

        # Extract features
        if self.preprocessing:
            features = self.preprocessing(audio)
        else:
            features = extract_features(audio)

        # Apply transforms if any
        if self.transform:
            features = self.transform(features)

        # Convert to tensor
        features_tensor = torch.tensor(features, dtype=torch.float32).unsqueeze(
            0
        )  # Add channel dimension

        return {
            "features": features_tensor,
            "label": torch.tensor(class_idx, dtype=torch.long),
            "file_name": file_name,
        }


def create_ssl_data_split(dataset, num_labeled_per_class=100, seed=42):
    """
    Split a dataset into labeled and unlabeled portions with a specific number of labeled samples per class.
    Returns both datasets to ensure no overlap.
    """
    random.seed(seed)

    # Get all indices and corresponding labels
    all_indices = list(range(len(dataset)))
    all_labels = [dataset[i]["label"].item() for i in all_indices]

    # Group indices by label
    label_indices = {}
    for idx, label in zip(all_indices, all_labels):
        if label not in label_indices:
            label_indices[label] = []
        label_indices[label].append(idx)

    # Select labeled indices
    labeled_indices = []
    unlabeled_indices = []

    for label, indices in label_indices.items():
        # Shuffle indices for randomness
        random.shuffle(indices)

        # Take first num_labeled_per_class as labeled
        if num_labeled_per_class <= len(indices):
            labeled_indices.extend(indices[:num_labeled_per_class])
            unlabeled_indices.extend(indices[num_labeled_per_class:])
        else:
            # If we don't have enough, take all as labeled
            labeled_indices.extend(indices)
            print(
                f"Warning: Class {label} has only {len(indices)} samples, less than requested {num_labeled_per_class}"
            )

    print(
        f"Created {len(labeled_indices)} labeled samples and {len(unlabeled_indices)} unlabeled samples"
    )

    # Create dataset subsets
    labeled_dataset = torch.utils.data.Subset(dataset, labeled_indices)
    unlabeled_dataset = torch.utils.data.Subset(dataset, unlabeled_indices)

    return labeled_dataset, unlabeled_dataset


def create_dataloaders(
    dataset, batch_size=32, num_labeled_per_class=None, num_workers=2
):
    if num_labeled_per_class is not None:
        # Split into labeled and unlabeled datasets
        labeled_dataset, unlabeled_dataset = create_ssl_data_split(
            dataset, num_labeled_per_class=num_labeled_per_class
        )

        # Ensure the unlabeled dataset is not empty
        if len(unlabeled_dataset) == 0:
            print(
                "Warning: No unlabeled data available. Using labeled data as placeholder."
            )
            # Create a small dummy unlabeled dataset from the labeled data (with different indices)
            unlabeled_dataset = torch.utils.data.Subset(
                dataset, list(range(min(len(dataset), 32)))
            )

        # Create dataloaders
        labeled_loader = DataLoader(
            labeled_dataset,
            batch_size=batch_size,
            shuffle=True,
            num_workers=num_workers,
        )

        unlabeled_loader = DataLoader(
            unlabeled_dataset,
            batch_size=batch_size * 2,  # Typically use larger batch size for unlabeled
            shuffle=True,
            num_workers=num_workers,
        )

        return labeled_loader, unlabeled_loader
    else:
        # If no labeled count is specified, return a single dataloader for evaluation
        return DataLoader(
            dataset, batch_size=batch_size, shuffle=True, num_workers=num_workers
        )


# -----------------------------------------------------
# Model Definition - MobileNetV2 for Audio Classification
# -----------------------------------------------------
class AudioClassifier(nn.Module):
    def __init__(self, num_classes, pretrained=True):
        super(AudioClassifier, self).__init__()

        # Load MobileNetV2 backbone
        self.mobilenet = models.mobilenet_v2(pretrained=pretrained)

        # Modify the first convolutional layer to accept 1-channel input (grayscale)
        # Keep pretrained weights for other channels, zero-initialize the new channel
        original_conv = self.mobilenet.features[0][0]
        self.mobilenet.features[0][0] = nn.Conv2d(
            1,
            original_conv.out_channels,
            kernel_size=original_conv.kernel_size,
            stride=original_conv.stride,
            padding=original_conv.padding,
            bias=original_conv.bias is not None,
        )

        # Initialize the new layer
        if pretrained:
            with torch.no_grad():
                new_weight = self.mobilenet.features[0][0].weight
                original_weight = original_conv.weight
                # Set new weights to average of original 3 channels
                new_weight[:, 0:1, :, :] = original_weight.mean(dim=1, keepdim=True)

        # Replace the classifier
        in_features = self.mobilenet.classifier[1].in_features
        self.mobilenet.classifier = nn.Sequential(
            nn.Dropout(0.2), nn.Linear(in_features, num_classes)
        )

    def forward(self, x):
        return self.mobilenet(x)

    def get_embedding(self, x):
        # Extract features before classification layer
        x = self.mobilenet.features(x)
        x = nn.functional.adaptive_avg_pool2d(x, (1, 1))
        x = torch.flatten(x, 1)
        return x


# -----------------------------------------------------
# FlatMatch Implementation
# -----------------------------------------------------
class FlatMatch:
    def __init__(
        self,
        model,
        optimizer,
        scheduler=None,
        num_classes=10,
        ema_decay=0.999,
        confidence_threshold=0.7,
        rampup_length=500,
        max_grad_norm=1.0,
        lambda_cs=1.0,
        rho=0.1,
        temperature=1.0,
        efficient=True,
    ):

        self.model = model
        self.optimizer = optimizer
        self.scheduler = scheduler
        self.num_classes = num_classes
        self.ema_decay = ema_decay
        self.confidence_threshold = confidence_threshold
        self.rampup_length = rampup_length
        self.max_grad_norm = max_grad_norm
        self.lambda_cs = lambda_cs
        self.rho = rho
        self.temperature = temperature
        self.efficient = efficient

        # Initialize memory buffer for storing historical gradients
        self.memory_buffer = None

        # Create EMA model
        self.ema = EMA(model, decay=ema_decay)

        # Store training metrics
        self.train_metrics = {
            "loss": [],
            "loss_sup": [],
            "loss_unsup": [],
            "cross_sharp": [],
            "mask_ratio": [],
            "learning_rate": [],
        }

    def sigmoid_rampup(self, current, rampup_length):
        """Exponential rampup from https://arxiv.org/abs/1610.02242"""
        if rampup_length == 0:
            return 1.0
        else:
            current = np.clip(current, 0.0, rampup_length)
            phase = 1.0 - current / rampup_length
            return float(np.exp(-5.0 * phase * phase))

    def compute_gradient(self, loss):
        """Compute gradient from loss"""
        self.optimizer.zero_grad()
        loss.backward(retain_graph=True)

        # Get gradients
        gradients = []
        for param in self.model.parameters():
            if param.grad is not None:
                gradients.append(param.grad.view(-1))

        gradients = torch.cat(gradients)
        return gradients

    def compute_optimal_perturbation(self, gradients):
        """Compute optimal perturbation based on gradients"""
        grad_norm = torch.norm(gradients, p=2)
        if grad_norm > 0:
            return self.rho * gradients / grad_norm
        return torch.zeros_like(gradients)

    def apply_perturbation(self, perturbation):
        """Apply perturbation to model parameters"""
        perturbed_params = []
        idx = 0
        for param in self.model.parameters():
            if param.requires_grad:
                param_size = param.numel()
                param_perturbation = perturbation[idx : idx + param_size].view(
                    param.size()
                )
                perturbed_params.append(param.data + param_perturbation)
                idx += param_size
            else:
                perturbed_params.append(param.data)

        return perturbed_params

    def restore_parameters(self, original_params):
        """Restore original parameters"""
        for param, orig in zip(self.model.parameters(), original_params):
            param.data.copy_(orig)

    def train_step(self, labeled_batch, unlabeled_batch, global_step):
        features_l = labeled_batch["features"].to(device)
        labels_l = labeled_batch["label"].to(device)

        features_u = unlabeled_batch["features"].to(device)

        # Calculate supervised loss
        logits_l = self.model(features_l)
        loss_sup = F.cross_entropy(logits_l, labels_l)

        # Save original parameters
        original_params = [p.data.clone() for p in self.model.parameters()]

        # Get pseudo-labels from the EMA model
        self.ema.apply_shadow()  # Apply EMA weights

        with torch.no_grad():
            logits_u_ema = self.model(features_u)
            probs_u_ema = F.softmax(logits_u_ema / self.temperature, dim=1)
            max_probs, pseudo_labels = torch.max(probs_u_ema, dim=1)

            # Create confidence mask (only use high-confidence predictions)
            # Use a lower initial threshold that increases over time
            threshold = max(
                0.5,
                self.confidence_threshold
                * self.sigmoid_rampup(global_step, self.rampup_length / 2),
            )
            mask = max_probs.ge(threshold).float()

            # Debug info to track pseudo-label statistics
            pseudo_label_count = mask.sum().item()
            mask_percentage = (
                100 * pseudo_label_count / len(mask) if len(mask) > 0 else 0
            )

            if global_step % 20 == 0:  # Print less frequently to reduce output
                print(
                    f"Step {global_step}: Generated {pseudo_label_count}/{len(mask)} pseudo-labels ({mask_percentage:.1f}%) with confidence >= {threshold:.4f}"
                )

        self.ema.restore()  # Restore original weights

        # Calculate unsupervised loss with pseudo-labels
        logits_u = self.model(features_u)

        # Calculate unsupervised loss
        if mask.sum() > 0:
            # Create one-hot encoded pseudo-labels for consistency loss
            pseudo_one_hot = F.one_hot(pseudo_labels, self.num_classes).float()

            # Compare model predictions with pseudo-labels using KL divergence
            probs_u = F.softmax(logits_u, dim=1)
            log_probs_u = F.log_softmax(logits_u, dim=1)
            loss_unsup_each = torch.sum(
                pseudo_one_hot * (torch.log(pseudo_one_hot + 1e-10) - log_probs_u),
                dim=1,
            )
            loss_unsup = (loss_unsup_each * mask).sum() / (mask.sum() + 1e-10)
        else:
            # If no confident samples, use a small entropy minimization loss
            probs_u = F.softmax(logits_u, dim=1)
            entropy = -torch.sum(probs_u * torch.log(probs_u + 1e-10), dim=1)
            loss_unsup = (
                0.05 * entropy.mean()
            )  # Small weight to avoid overwhelming supervised loss

        # Calculate weight for unsupervised loss (ramp-up)
        w = self.sigmoid_rampup(global_step, self.rampup_length)

        # Compute gradient for adversarial perturbation
        if self.efficient:
            # FlatMatch-e: use only supervised loss for perturbation
            gradients = self.compute_gradient(loss_sup)
        else:
            # Standard FlatMatch: use both losses
            combined_loss = loss_sup + w * loss_unsup
            gradients = self.compute_gradient(combined_loss)

        # Compute and apply perturbation
        perturbation = self.compute_optimal_perturbation(gradients)
        perturbed_params = self.apply_perturbation(perturbation)

        # Store original parameters
        orig_params = [p.data.clone() for p in self.model.parameters()]

        # Apply perturbed parameters
        for param, perturbed in zip(self.model.parameters(), perturbed_params):
            if param.requires_grad:
                param.data.copy_(perturbed)

        # Calculate cross-sharpness term
        logits_l_perturbed = self.model(features_l)
        loss_l_perturbed = F.cross_entropy(logits_l_perturbed, labels_l)

        logits_u_perturbed = self.model(features_u)

        # Calculate perturbed unsupervised loss with the same method as before
        if mask.sum() > 0:
            probs_u_perturbed = F.softmax(logits_u_perturbed, dim=1)
            log_probs_u_perturbed = F.log_softmax(logits_u_perturbed, dim=1)
            loss_u_perturbed_each = torch.sum(
                pseudo_one_hot
                * (torch.log(pseudo_one_hot + 1e-10) - log_probs_u_perturbed),
                dim=1,
            )
            loss_u_perturbed = (loss_u_perturbed_each * mask).sum() / (
                mask.sum() + 1e-10
            )
        else:
            loss_u_perturbed = torch.tensor(0.0, device=device)

        # Cross-sharpness is the sum of both perturbed losses
        cross_sharp = loss_l_perturbed + loss_u_perturbed

        # Restore original parameters
        for param, orig in zip(self.model.parameters(), orig_params):
            param.data.copy_(orig)

        # Calculate total loss
        total_loss = loss_sup + w * loss_unsup + self.lambda_cs * cross_sharp

        # Optimize model parameters
        self.optimizer.zero_grad()
        total_loss.backward()

        if self.max_grad_norm > 0:
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), self.max_grad_norm)

        self.optimizer.step()

        # Update EMA model
        self.ema.update()

        # Update scheduler
        if self.scheduler is not None:
            self.scheduler.step()
            current_lr = self.scheduler.get_last_lr()[0]
        else:
            current_lr = self.optimizer.param_groups[0]["lr"]

        # Store metrics for later analysis
        metrics = {
            "loss": total_loss.item(),
            "loss_sup": loss_sup.item(),
            "loss_unsup": loss_unsup.item() * w,
            "cross_sharp": cross_sharp.item(),
            "mask_ratio": mask.mean().item(),
            "learning_rate": current_lr,
        }

        for key, value in metrics.items():
            self.train_metrics[key].append(value)

        return metrics


# -----------------------------------------------------
# Training and Evaluation Functions
# -----------------------------------------------------
def evaluate(model, dataloader, num_classes):
    model.eval()
    all_preds = []
    all_labels = []
    total_loss = 0

    with torch.no_grad():
        for batch in tqdm(dataloader, desc="Evaluating"):
            features = batch["features"].to(device)
            labels = batch["label"].to(device)

            logits = model(features)
            loss = F.cross_entropy(logits, labels)
            total_loss += loss.item()

            preds = torch.argmax(logits, dim=1).cpu().numpy()
            all_preds.extend(preds)
            all_labels.extend(labels.cpu().numpy())

    # Calculate metrics
    accuracy = accuracy_score(all_labels, all_preds)
    precision = precision_score(all_labels, all_preds, average="macro", zero_division=0)
    recall = recall_score(all_labels, all_preds, average="macro", zero_division=0)
    f1 = f1_score(all_labels, all_preds, average="macro", zero_division=0)

    # Calculate confusion matrix
    cm = confusion_matrix(all_labels, all_preds, labels=range(num_classes))

    avg_loss = total_loss / len(dataloader)

    return {
        "loss": avg_loss,
        "accuracy": accuracy,
        "precision": precision,
        "recall": recall,
        "f1": f1,
        "confusion_matrix": cm,
        "predictions": all_preds,
        "true_labels": all_labels,
    }


def plot_confusion_matrix(cm, class_names, title, filepath):
    """Plot confusion matrix with proper normalization."""
    plt.figure(figsize=(12, 10))

    # Normalize by row (true labels)
    cm_normalized = cm.astype("float") / (cm.sum(axis=1)[:, np.newaxis] + 1e-10)

    sns.heatmap(
        cm_normalized,
        annot=True,
        fmt=".2f",
        cmap="Blues",
        xticklabels=class_names,
        yticklabels=class_names,
    )

    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title(title)
    plt.tight_layout()
    plt.savefig(filepath)
    plt.close()


def plot_training_metrics(train_metrics, filepath_prefix):
    """Plot training metrics."""
    for metric_name, values in train_metrics.items():
        if len(values) == 0:
            continue

        plt.figure(figsize=(10, 6))
        plt.plot(values)
        plt.xlabel("Training Steps")
        plt.ylabel(metric_name.replace("_", " ").title())
        plt.title(f'{metric_name.replace("_", " ").title()} During Training')
        plt.grid(True)
        plt.savefig(f"{filepath_prefix}_{metric_name}.png")
        plt.close()


def train_and_evaluate(
    labeled_loader,
    unlabeled_loader,
    val_loader,
    test_loader,
    num_labeled,
    num_classes,
    epochs=10,
    experiment_name="default",
    efficient=True,
):

    # Create model
    model = AudioClassifier(num_classes=num_classes, pretrained=True).to(device)

    # Setup optimizer and scheduler
    optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-4)
    total_steps = epochs * len(labeled_loader)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=total_steps)

    # Setup FlatMatch
    flatmatch = FlatMatch(
        model=model,
        optimizer=optimizer,
        scheduler=scheduler,
        num_classes=num_classes,
        confidence_threshold=0.7,
        rampup_length=int(0.4 * total_steps),
        lambda_cs=1.0,
        rho=0.05,
        temperature=1.0,
        efficient=efficient,
    )

    # Training loop
    global_step = 0
    train_losses = []
    val_losses = []
    val_metrics_history = []
    best_val_accuracy = 0
    best_model_path = os.path.join(
        OUTPUT_DIR, f"best_model_{experiment_name}_{num_labeled}.pt"
    )

    # Early stopping parameters
    patience = 3
    patience_counter = 0

    for epoch in range(epochs):
        model.train()
        epoch_losses = []

        # Create iterator for the unlabeled dataloader
        unlabeled_iter = iter(unlabeled_loader)

        # Training
        pbar = tqdm(labeled_loader, desc=f"Epoch {epoch+1}/{epochs}")
        for labeled_batch in pbar:
            # Get unlabeled batch
            try:
                unlabeled_batch = next(unlabeled_iter)
            except StopIteration:
                unlabeled_iter = iter(unlabeled_loader)
                unlabeled_batch = next(unlabeled_iter)

            # Train step
            step_results = flatmatch.train_step(
                labeled_batch, unlabeled_batch, global_step
            )
            epoch_losses.append(step_results["loss"])

            # Display progress
            pbar.set_postfix(
                loss=f"{step_results['loss']:.4f}",
                sup_loss=f"{step_results['loss_sup']:.4f}",
                unsup_loss=f"{step_results['loss_unsup']:.4f}",
                mask=f"{step_results['mask_ratio']:.2f}",
            )

            global_step += 1

        # Evaluation on validation set
        val_metrics = evaluate(model, val_loader, num_classes)
        val_metrics_history.append(val_metrics)

        # Track losses
        avg_train_loss = np.mean(epoch_losses)
        train_losses.append(avg_train_loss)
        val_losses.append(val_metrics["loss"])

        # Save the best model
        if val_metrics["accuracy"] > best_val_accuracy:
            best_val_accuracy = val_metrics["accuracy"]
            torch.save(model.state_dict(), best_model_path)
            patience_counter = 0
        else:
            patience_counter += 1

        print(
            f"\nEpoch {epoch+1}/{epochs} - Train Loss: {avg_train_loss:.4f}, Val Loss: {val_metrics['loss']:.4f}, "
            f"Val Accuracy: {val_metrics['accuracy']:.4f}, Val F1: {val_metrics['f1']:.4f}"
        )

        # Early stopping
        if patience_counter >= patience:
            print(f"Early stopping triggered after {epoch+1} epochs")
            break

    # Plot training and validation loss
    plt.figure(figsize=(10, 6))
    plt.plot(train_losses, label="Training Loss")
    plt.plot(val_losses, label="Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(
        f'Training and Validation Loss (Labeled Samples: {num_labeled}, {"FlatMatch-e" if efficient else "FlatMatch"})'
    )
    plt.legend()
    plt.grid(True)
    plt.savefig(
        os.path.join(OUTPUT_DIR, f"loss_curve_{experiment_name}_{num_labeled}.png")
    )

    # Plot training metrics
    # Continuing from where the code left off...

    # Plot training metrics
    plot_training_metrics(
        flatmatch.train_metrics,
        os.path.join(OUTPUT_DIR, f"metrics_{experiment_name}_{num_labeled}"),
    )

    # Load best model for evaluation
    model.load_state_dict(torch.load(best_model_path))
    model.eval()

    # Evaluate on test set
    test_metrics = evaluate(model, test_loader, num_classes)

    # Plot confusion matrix
    plot_confusion_matrix(
        test_metrics["confusion_matrix"],
        val_loader.dataset.dataset.class_names
        if hasattr(val_loader.dataset, "dataset")
        else val_loader.dataset.class_names,
        f"Confusion Matrix (Labeled: {num_labeled})",
        os.path.join(
            OUTPUT_DIR, f"confusion_matrix_{experiment_name}_{num_labeled}.png"
        ),
    )

    # Save results to CSV
    results_df = pd.DataFrame(
        {
            "experiment": experiment_name,
            "num_labeled": num_labeled,
            "accuracy": test_metrics["accuracy"],
            "precision": test_metrics["precision"],
            "recall": test_metrics["recall"],
            "f1": test_metrics["f1"],
            "val_accuracy": best_val_accuracy,
        },
        index=[0],
    )

    results_path = os.path.join(
        OUTPUT_DIR, f"results_{experiment_name}_{num_labeled}.csv"
    )
    results_df.to_csv(results_path, index=False)

    return {
        "model": model,
        "train_losses": train_losses,
        "val_losses": val_losses,
        "val_metrics_history": val_metrics_history,
        "test_metrics": test_metrics,
        "results_df": results_df,
    }


# -----------------------------------------------------
# Main Experiment
# -----------------------------------------------------
def main():
    # Load datasets
    # Set parameters for data preprocessing
    sample_rate = 22050
    n_mels = 128
    duration = 5  # seconds

    print("Loading datasets...")

    # Helper function for preprocessing
    def audio_preprocessing(audio):
        return extract_features(audio, sample_rate=sample_rate, n_mels=n_mels)

    # Load training dataset
    train_dataset = AudioDataset(
        TRAIN_METADATA, TRAIN_PATH, preprocessing=audio_preprocessing
    )

    # Load validation dataset
    val_dataset = AudioDataset(
        VAL_METADATA, VAL_PATH, preprocessing=audio_preprocessing
    )

    # Load test dataset
    test_dataset = AudioDataset(
        TEST_METADATA, TEST_PATH, preprocessing=audio_preprocessing
    )

    print(
        f"Loaded {len(train_dataset)} training samples, {len(val_dataset)} validation samples, and {len(test_dataset)} test samples"
    )

    # Number of classes
    num_classes = len(train_dataset.class_names)
    print(f"Number of classes: {num_classes}")

    # Use the same class names for all datasets
    val_dataset.class_to_idx = train_dataset.class_to_idx
    val_dataset.class_names = train_dataset.class_names
    test_dataset.class_to_idx = train_dataset.class_to_idx
    test_dataset.class_names = train_dataset.class_names

    # Create validation and test dataloaders (no SSL split)
    val_loader = create_dataloaders(val_dataset, batch_size=32, num_workers=2)
    test_loader = create_dataloaders(test_dataset, batch_size=32, num_workers=2)

    # Define the labeled samples per class to experiment with
    labeled_per_class_options = [40, 100, 250, 500]

    # Dictionary to store all results
    all_results = {}
    all_results_df = []

    # Run experiments for each number of labeled samples
    for labeled_per_class in labeled_per_class_options:
        print(f"\n{'='*80}")
        print(f"Running experiment with {labeled_per_class} labeled samples per class")
        print(f"{'='*80}")

        # Create dataloaders with the specific number of labeled samples
        labeled_loader, unlabeled_loader = create_dataloaders(
            train_dataset,
            batch_size=32,
            num_labeled_per_class=labeled_per_class,
            num_workers=2,
        )

        # Calculate total labeled samples
        total_labeled = len(labeled_loader.dataset)

        print(
            f"Created dataloaders with {total_labeled} labeled samples and {len(unlabeled_loader.dataset)} unlabeled samples."
        )

        # Run FlatMatch-e (efficient version)
        print("\nTraining with FlatMatch-e (efficient version)...")
        flatmatch_e_results = train_and_evaluate(
            labeled_loader,
            unlabeled_loader,
            val_loader,
            test_loader,
            labeled_per_class,
            num_classes,
            epochs=30,
            experiment_name="FlatMatch_e",
            efficient=True,
        )

        # Run standard FlatMatch
        print("\nTraining with standard FlatMatch...")
        flatmatch_results = train_and_evaluate(
            labeled_loader,
            unlabeled_loader,
            val_loader,
            test_loader,
            labeled_per_class,
            num_classes,
            epochs=30,
            experiment_name="FlatMatch",
            efficient=False,
        )

        # Store results
        all_results[f"flatmatch_e_{labeled_per_class}"] = flatmatch_e_results
        all_results[f"flatmatch_{labeled_per_class}"] = flatmatch_results

        # Concatenate result DataFrames
        all_results_df.append(flatmatch_e_results["results_df"])
        all_results_df.append(flatmatch_results["results_df"])

    # Combine all results into a single DataFrame
    combined_results = pd.concat(all_results_df, ignore_index=True)
    combined_results.to_csv(os.path.join(OUTPUT_DIR, "all_results.csv"), index=False)

    # Plot comparison of methods for different amounts of labeled data
    plt.figure(figsize=(12, 8))

    # Get unique experiment names
    experiment_names = combined_results["experiment"].unique()

    # Plot line for each experiment
    for experiment in experiment_names:
        subset = combined_results[combined_results["experiment"] == experiment]
        plt.plot(
            subset["num_labeled"], subset["accuracy"], marker="o", label=experiment
        )

    plt.xlabel("Number of Labeled Examples per Class")
    plt.ylabel("Test Accuracy")
    plt.title("Performance Comparison with Different Amounts of Labeled Data")
    plt.grid(True)
    plt.legend()
    plt.savefig(os.path.join(OUTPUT_DIR, "method_comparison.png"))
    plt.close()

    # Print final results table
    print("\n===== Final Results =====")
    print(combined_results.to_string(index=False))

    print(f"\nAll results saved to {OUTPUT_DIR}")


if __name__ == "__main__":
    main()
```
