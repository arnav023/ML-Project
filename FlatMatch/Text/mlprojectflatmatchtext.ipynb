```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
)
from transformers import BertTokenizer, BertModel
import seaborn as sns
import random
from tqdm import tqdm

# Set seeds for reproducibility
random.seed(42)
np.random.seed(42)
torch.manual_seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(42)

# Create output directory
OUTPUT_DIR = "Flatmatch_TextData"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Define device
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Number of classes in Yahoo Answers dataset
NUM_CLASSES = 10


# -----------------------------------------------------
# EMA Implementation
# -----------------------------------------------------
class EMA:
    def __init__(self, model, decay):
        self.model = model
        self.decay = decay
        self.shadow = {}
        self.backup = {}
        self.register()

    def register(self):
        for name, param in self.model.named_parameters():
            if param.requires_grad:
                self.shadow[name] = param.data.clone()

    def update(self):
        for name, param in self.model.named_parameters():
            if param.requires_grad:
                assert name in self.shadow
                new_average = (
                    1.0 - self.decay
                ) * param.data + self.decay * self.shadow[name]
                self.shadow[name] = new_average.clone()

    def apply_shadow(self):
        for name, param in self.model.named_parameters():
            if param.requires_grad:
                assert name in self.shadow
                self.backup[name] = param.data.clone()
                param.data = self.shadow[name]

    def restore(self):
        for name, param in self.model.named_parameters():
            if param.requires_grad:
                assert name in self.backup
                param.data = self.backup[name]
        self.backup = {}


# -----------------------------------------------------
# Data Loading and Preprocessing
# -----------------------------------------------------
class YahooAnswersDataset(Dataset):
    def __init__(
        self,
        data_dir,
        tokenizer,
        max_length=128,
        transform=None,
        num_samples_per_class=None,
        max_files=1000,
    ):
        self.data_dir = data_dir
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.transform = transform

        self.texts = []
        self.labels = []

        class_dirs = sorted(
            [
                d
                for d in os.listdir(data_dir)
                if os.path.isdir(os.path.join(data_dir, d))
            ]
        )

        for label, class_dir in enumerate(class_dirs):
            class_path = os.path.join(data_dir, class_dir)
            file_paths = [
                os.path.join(class_path, f)
                for f in os.listdir(class_path)
                if f.endswith(".txt")
            ]

            # Limit to max_files per class
            if max_files and len(file_paths) > max_files:
                file_paths = file_paths[:max_files]

            # Further limit if num_samples_per_class is specified
            if num_samples_per_class and num_samples_per_class < len(file_paths):
                file_paths = random.sample(file_paths, num_samples_per_class)

            print(f"Loading class {class_dir} with {len(file_paths)} files")

            for file_path in file_paths:
                with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
                    text = f.read().strip()

                self.texts.append(text)
                self.labels.append(label)

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        label = self.labels[idx]

        encoding = self.tokenizer(
            text,
            truncation=True,
            max_length=self.max_length,
            padding="max_length",
            return_tensors="pt",
        )

        # Convert to appropriate format for model
        input_ids = encoding["input_ids"].squeeze(0)
        attention_mask = encoding["attention_mask"].squeeze(0)

        return {
            "input_ids": input_ids,
            "attention_mask": attention_mask,
            "label": torch.tensor(label, dtype=torch.long),
        }


def create_ssl_data_split(dataset, num_labeled_per_class=100):
    """
    Split a dataset into labeled and unlabeled portions with a specific number of labeled samples per class.
    Returns both datasets to ensure no overlap.
    """
    # Group indices by label
    label_indices = {}
    for idx, sample in enumerate(dataset):
        label = sample["label"].item()
        if label not in label_indices:
            label_indices[label] = []
        label_indices[label].append(idx)

    # Select labeled indices
    labeled_indices = []
    unlabeled_indices = []

    for label, indices in label_indices.items():
        # Shuffle indices for randomness
        random.shuffle(indices)

        # Take first num_labeled_per_class as labeled
        if num_labeled_per_class < len(indices):
            labeled_indices.extend(indices[:num_labeled_per_class])
            unlabeled_indices.extend(indices[num_labeled_per_class:])
        else:
            # If we don't have enough, take all as labeled
            labeled_indices.extend(indices)

    # Create dataset subsets
    labeled_dataset = torch.utils.data.Subset(dataset, labeled_indices)
    unlabeled_dataset = torch.utils.data.Subset(dataset, unlabeled_indices)

    return labeled_dataset, unlabeled_dataset


def create_dataloaders(
    data_dir,
    tokenizer,
    max_length=128,
    batch_size=32,
    num_labeled_per_class=None,
    max_files=1000,
):
    # Create full dataset
    full_dataset = YahooAnswersDataset(
        data_dir, tokenizer, max_length, max_files=max_files
    )

    if num_labeled_per_class is not None:
        # Split into labeled and unlabeled datasets
        labeled_dataset, unlabeled_dataset = create_ssl_data_split(
            full_dataset, num_labeled_per_class=num_labeled_per_class
        )

        # Create dataloaders
        labeled_loader = DataLoader(
            labeled_dataset, batch_size=batch_size, shuffle=True, num_workers=2
        )

        unlabeled_loader = DataLoader(
            unlabeled_dataset,
            batch_size=batch_size * 2,  # Typically use larger batch size for unlabeled
            shuffle=True,
            num_workers=2,
        )

        return labeled_loader, unlabeled_loader
    else:
        # If no labeled count is specified, return a single dataloader for evaluation
        return DataLoader(
            full_dataset, batch_size=batch_size, shuffle=True, num_workers=2
        )


# -----------------------------------------------------
# Model Definition
# -----------------------------------------------------
class TextClassifier(nn.Module):
    def __init__(self, num_classes=NUM_CLASSES):
        super(TextClassifier, self).__init__()
        self.bert = BertModel.from_pretrained("bert-base-uncased")
        self.dropout = nn.Dropout(0.1)
        self.fc = nn.Linear(768, num_classes)

    def forward(self, input_ids, attention_mask):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        pooled_output = outputs.pooler_output
        x = self.dropout(pooled_output)
        logits = self.fc(x)
        return logits

    def get_embedding(self, input_ids, attention_mask):
        with torch.no_grad():
            outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
            return outputs.pooler_output


# -----------------------------------------------------
# FlatMatch Implementation
# -----------------------------------------------------
class FlatMatch:
    def __init__(
        self,
        model,
        optimizer,
        scheduler=None,
        num_classes=NUM_CLASSES,
        ema_decay=0.999,
        confidence_threshold=0.7,
        rampup_length=500,
        max_grad_norm=1.0,
        lambda_cs=1.0,
        rho=0.1,
        temperature=1.0,
        efficient=True,
    ):

        self.model = model
        self.optimizer = optimizer
        self.scheduler = scheduler
        self.num_classes = num_classes
        self.ema_decay = ema_decay
        self.confidence_threshold = (
            confidence_threshold  # Lower threshold to get more pseudo-labels
        )
        self.rampup_length = rampup_length
        self.max_grad_norm = max_grad_norm
        self.lambda_cs = lambda_cs
        self.rho = rho
        self.temperature = temperature  # Changed to not sharpen too much
        self.efficient = efficient

        # Initialize memory buffer for storing historical gradients
        self.memory_buffer = None

        # Create EMA model
        self.ema = EMA(model, decay=ema_decay)

        # Store training metrics
        self.train_metrics = {
            "loss": [],
            "loss_sup": [],
            "loss_unsup": [],
            "cross_sharp": [],
            "mask_ratio": [],
            "learning_rate": [],
        }

    def sigmoid_rampup(self, current, rampup_length):
        """Exponential rampup from https://arxiv.org/abs/1610.02242"""
        if rampup_length == 0:
            return 1.0
        else:
            current = np.clip(current, 0.0, rampup_length)
            phase = 1.0 - current / rampup_length
            return float(np.exp(-5.0 * phase * phase))

    def compute_gradient(self, loss):
        """Compute gradient from loss"""
        self.optimizer.zero_grad()
        loss.backward(retain_graph=True)

        # Get gradients
        gradients = []
        for param in self.model.parameters():
            if param.grad is not None:
                gradients.append(param.grad.view(-1))

        gradients = torch.cat(gradients)
        return gradients

    def compute_optimal_perturbation(self, gradients):
        """Compute optimal perturbation based on gradients"""
        grad_norm = torch.norm(gradients, p=2)
        if grad_norm > 0:
            return self.rho * gradients / grad_norm
        return torch.zeros_like(gradients)

    def apply_perturbation(self, perturbation):
        """Apply perturbation to model parameters"""
        perturbed_params = []
        idx = 0
        for param in self.model.parameters():
            if param.requires_grad:
                param_size = param.numel()
                param_perturbation = perturbation[idx : idx + param_size].view(
                    param.size()
                )
                perturbed_params.append(param.data + param_perturbation)
                idx += param_size
            else:
                perturbed_params.append(param.data)

        return perturbed_params

    def restore_parameters(self, original_params):
        """Restore original parameters"""
        for param, orig in zip(self.model.parameters(), original_params):
            param.data.copy_(orig)

    def train_step(self, labeled_batch, unlabeled_batch, global_step):
        input_ids_l = labeled_batch["input_ids"].to(device)
        attention_mask_l = labeled_batch["attention_mask"].to(device)
        labels_l = labeled_batch["label"].to(device)

        input_ids_u = unlabeled_batch["input_ids"].to(device)
        attention_mask_u = unlabeled_batch["attention_mask"].to(device)

        # Calculate supervised loss
        logits_l = self.model(input_ids_l, attention_mask_l)
        loss_sup = F.cross_entropy(logits_l, labels_l)

        # Save original parameters
        original_params = [p.data.clone() for p in self.model.parameters()]

        # Get pseudo-labels from the EMA model
        self.ema.apply_shadow()  # Apply EMA weights

        with torch.no_grad():
            logits_u_ema = self.model(input_ids_u, attention_mask_u)
            probs_u_ema = F.softmax(logits_u_ema / self.temperature, dim=1)
            max_probs, pseudo_labels = torch.max(probs_u_ema, dim=1)

            # Create confidence mask (only use high-confidence predictions)
            # Use a lower initial threshold that increases over time
            threshold = max(
                0.5,
                self.confidence_threshold
                * self.sigmoid_rampup(global_step, self.rampup_length / 2),
            )
            mask = max_probs.ge(threshold).float()

            # Debug info to track pseudo-label statistics
            pseudo_label_count = mask.sum().item()
            if pseudo_label_count > 0:
                print(
                    f"Step {global_step}: Generated {pseudo_label_count}/{len(mask)} pseudo-labels with confidence >= {threshold:.4f}"
                )
            else:
                print(
                    f"Step {global_step}: No pseudo-labels generated (threshold: {threshold:.4f})"
                )

        self.ema.restore()  # Restore original weights

        # Calculate unsupervised loss with pseudo-labels
        logits_u = self.model(input_ids_u, attention_mask_u)

        # Calculate unsupervised loss - THIS IS THE FIXED PART
        # The problem was with the use of cross_entropy and the mask application
        if mask.sum() > 0:
            # Create one-hot encoded pseudo-labels for consistency loss
            pseudo_one_hot = F.one_hot(pseudo_labels, self.num_classes).float()

            # Compare model predictions with pseudo-labels using KL divergence
            # This avoids the zero loss issue by focusing on distribution alignment
            probs_u = F.softmax(logits_u, dim=1)
            log_probs_u = F.log_softmax(logits_u, dim=1)
            loss_unsup_each = torch.sum(
                pseudo_one_hot * (torch.log(pseudo_one_hot + 1e-10) - log_probs_u),
                dim=1,
            )
            loss_unsup = (loss_unsup_each * mask).sum() / (mask.sum() + 1e-10)
        else:
            # If no confident samples, use a small entropy minimization loss
            # to keep the model learning from unlabeled data
            probs_u = F.softmax(logits_u, dim=1)
            entropy = -torch.sum(probs_u * torch.log(probs_u + 1e-10), dim=1)
            loss_unsup = (
                0.05 * entropy.mean()
            )  # Small weight to avoid overwhelming supervised loss

        # Calculate weight for unsupervised loss (ramp-up)
        w = self.sigmoid_rampup(global_step, self.rampup_length)

        # Compute gradient for adversarial perturbation
        if self.efficient:
            # FlatMatch-e: use only supervised loss for perturbation
            gradients = self.compute_gradient(loss_sup)
        else:
            # Standard FlatMatch: use both losses
            combined_loss = loss_sup + w * loss_unsup
            gradients = self.compute_gradient(combined_loss)

        # Compute and apply perturbation
        perturbation = self.compute_optimal_perturbation(gradients)
        perturbed_params = self.apply_perturbation(perturbation)

        # Store original parameters
        orig_params = [p.data.clone() for p in self.model.parameters()]

        # Apply perturbed parameters
        for param, perturbed in zip(self.model.parameters(), perturbed_params):
            if param.requires_grad:
                param.data.copy_(perturbed)

        # Calculate cross-sharpness term
        logits_l_perturbed = self.model(input_ids_l, attention_mask_l)
        loss_l_perturbed = F.cross_entropy(logits_l_perturbed, labels_l)

        logits_u_perturbed = self.model(input_ids_u, attention_mask_u)

        # Calculate perturbed unsupervised loss with the same method as before
        if mask.sum() > 0:
            probs_u_perturbed = F.softmax(logits_u_perturbed, dim=1)
            log_probs_u_perturbed = F.log_softmax(logits_u_perturbed, dim=1)
            loss_u_perturbed_each = torch.sum(
                pseudo_one_hot
                * (torch.log(pseudo_one_hot + 1e-10) - log_probs_u_perturbed),
                dim=1,
            )
            loss_u_perturbed = (loss_u_perturbed_each * mask).sum() / (
                mask.sum() + 1e-10
            )
        else:
            loss_u_perturbed = torch.tensor(0.0, device=device)

        # Cross-sharpness is the sum of both perturbed losses
        cross_sharp = loss_l_perturbed + loss_u_perturbed

        # Restore original parameters
        for param, orig in zip(self.model.parameters(), orig_params):
            param.data.copy_(orig)

        # Calculate total loss
        total_loss = loss_sup + w * loss_unsup + self.lambda_cs * cross_sharp

        # Optimize model parameters
        self.optimizer.zero_grad()
        total_loss.backward()

        if self.max_grad_norm > 0:
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), self.max_grad_norm)

        self.optimizer.step()

        # Update EMA model
        self.ema.update()

        # Update scheduler
        if self.scheduler is not None:
            self.scheduler.step()
            current_lr = self.scheduler.get_last_lr()[0]
        else:
            current_lr = self.optimizer.param_groups[0]["lr"]

        # Store metrics for later analysis
        metrics = {
            "loss": total_loss.item(),
            "loss_sup": loss_sup.item(),
            "loss_unsup": loss_unsup.item() * w,
            "cross_sharp": cross_sharp.item(),
            "mask_ratio": mask.mean().item(),
            "learning_rate": current_lr,
        }

        for key, value in metrics.items():
            self.train_metrics[key].append(value)

        return metrics


# -----------------------------------------------------
# Training and Evaluation Functions
# -----------------------------------------------------
def evaluate(model, dataloader, num_classes=NUM_CLASSES):
    model.eval()
    all_preds = []
    all_labels = []
    total_loss = 0

    with torch.no_grad():
        for batch in tqdm(dataloader, desc="Evaluating"):
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            labels = batch["label"].to(device)

            logits = model(input_ids, attention_mask)
            loss = F.cross_entropy(logits, labels)
            total_loss += loss.item()

            preds = torch.argmax(logits, dim=1).cpu().numpy()
            all_preds.extend(preds)
            all_labels.extend(labels.cpu().numpy())

    # Calculate metrics
    accuracy = accuracy_score(all_labels, all_preds)
    precision = precision_score(all_labels, all_preds, average="macro")
    recall = recall_score(all_labels, all_preds, average="macro")
    f1 = f1_score(all_labels, all_preds, average="macro")

    # Calculate confusion matrix
    cm = confusion_matrix(all_labels, all_preds, labels=range(num_classes))

    avg_loss = total_loss / len(dataloader)

    return {
        "loss": avg_loss,
        "accuracy": accuracy,
        "precision": precision,
        "recall": recall,
        "f1": f1,
        "confusion_matrix": cm,
        "predictions": all_preds,
        "true_labels": all_labels,
    }


def plot_confusion_matrix(cm, class_names, title, filepath):
    """Plot confusion matrix with proper normalization."""
    plt.figure(figsize=(10, 8))

    # Normalize by row (true labels)
    cm_normalized = cm.astype("float") / cm.sum(axis=1)[:, np.newaxis]

    sns.heatmap(
        cm_normalized,
        annot=True,
        fmt=".2f",
        cmap="Blues",
        xticklabels=class_names,
        yticklabels=class_names,
    )

    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title(title)
    plt.tight_layout()
    plt.savefig(filepath)
    plt.close()


def plot_training_metrics(train_metrics, filepath_prefix):
    """Plot training metrics."""
    for metric_name, values in train_metrics.items():
        if len(values) == 0:
            continue

        plt.figure(figsize=(10, 6))
        plt.plot(values)
        plt.xlabel("Training Steps")
        plt.ylabel(metric_name.replace("_", " ").title())
        plt.title(f'{metric_name.replace("_", " ").title()} During Training')
        plt.grid(True)
        plt.savefig(f"{filepath_prefix}_{metric_name}.png")
        plt.close()


def train_and_evaluate(
    labeled_loader,
    unlabeled_loader,
    val_loader,
    test_loader,
    num_labeled,
    epochs=10,
    experiment_name="default",
    efficient=True,
):

    # Create model
    model = TextClassifier(num_classes=NUM_CLASSES).to(device)

    # Setup optimizer and scheduler
    optimizer = optim.AdamW(model.parameters(), lr=2e-5, weight_decay=1e-4)
    total_steps = epochs * len(labeled_loader)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=total_steps)

    # Setup FlatMatch
    flatmatch = FlatMatch(
        model=model,
        optimizer=optimizer,
        scheduler=scheduler,
        confidence_threshold=0.7,  # Lower threshold to get more pseudo-labels
        rampup_length=int(0.4 * total_steps),  # Ramp up over 40% of training
        lambda_cs=1.0,
        rho=0.05,  # Smaller perturbation for text data
        temperature=1.0,  # Use standard temperature (not sharpening)
        efficient=efficient,
    )

    # Training loop
    global_step = 0
    train_losses = []
    val_losses = []
    val_metrics_history = []
    best_val_accuracy = 0
    best_model_path = os.path.join(
        OUTPUT_DIR, f"best_model_{experiment_name}_{num_labeled}.pt"
    )

    # Early stopping parameters
    patience = 3
    patience_counter = 0

    for epoch in range(epochs):
        model.train()
        epoch_losses = []

        # Create iterator for the unlabeled dataloader
        unlabeled_iter = iter(unlabeled_loader)

        # Training
        pbar = tqdm(labeled_loader, desc=f"Epoch {epoch+1}/{epochs}")
        for labeled_batch in pbar:
            # Get unlabeled batch
            try:
                unlabeled_batch = next(unlabeled_iter)
            except StopIteration:
                unlabeled_iter = iter(unlabeled_loader)
                unlabeled_batch = next(unlabeled_iter)

            # Train step
            step_results = flatmatch.train_step(
                labeled_batch, unlabeled_batch, global_step
            )
            epoch_losses.append(step_results["loss"])

            # Display unsup_loss with higher precision to check it's not exactly zero
            pbar.set_postfix(
                loss=f"{step_results['loss']:.4f}",
                sup_loss=f"{step_results['loss_sup']:.12f}",
                unsup_loss=f"{step_results['loss_unsup']:.12f}",
                mask=f"{step_results['mask_ratio']:.2f}",
            )

            global_step += 1

        # Evaluation on validation set
        val_metrics = evaluate(model, val_loader)
        val_metrics_history.append(val_metrics)

        # Track losses
        avg_train_loss = np.mean(epoch_losses)
        train_losses.append(avg_train_loss)
        val_losses.append(val_metrics["loss"])

        # Save the best model
        if val_metrics["accuracy"] > best_val_accuracy:
            best_val_accuracy = val_metrics["accuracy"]
            torch.save(model.state_dict(), best_model_path)
            patience_counter = 0
        else:
            patience_counter += 1

        print(
            f"\nEpoch {epoch+1}/{epochs} - Train Loss: {avg_train_loss:.4f}, Val Loss: {val_metrics['loss']:.4f}, "
            f"Val Accuracy: {val_metrics['accuracy']:.4f}, Val F1: {val_metrics['f1']:.4f}"
        )

        # Early stopping
        if patience_counter >= patience:
            print(f"Early stopping triggered after {epoch+1} epochs")
            break

    # Plot training and validation loss
    plt.figure(figsize=(10, 6))
    plt.plot(train_losses, label="Training Loss")
    plt.plot(val_losses, label="Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(
        f'Training and Validation Loss (Labeled Samples: {num_labeled}, {"FlatMatch-e" if efficient else "FlatMatch"})'
    )
    plt.legend()
    plt.grid(True)
    plt.savefig(
        os.path.join(OUTPUT_DIR, f"loss_curve_{experiment_name}_{num_labeled}.png")
    )

    # Plot training metrics
    metrics_filepath_prefix = os.path.join(
        OUTPUT_DIR, f"{experiment_name}_{num_labeled}"
    )
    plot_training_metrics(flatmatch.train_metrics, metrics_filepath_prefix)

    # Load the best model for final evaluation
    model.load_state_dict(torch.load(best_model_path))
    test_metrics = evaluate(model, test_loader)

    # Plot confusion matrix
    class_names = [str(i) for i in range(NUM_CLASSES)]
    plot_confusion_matrix(
        test_metrics["confusion_matrix"],
        class_names,
        f"Confusion Matrix - {experiment_name} ({num_labeled} samples per class)",
        os.path.join(
            OUTPUT_DIR, f"confusion_matrix_{experiment_name}_{num_labeled}.png"
        ),
    )

    print(
        f"Test Results (Labeled Samples: {num_labeled}, {'FlatMatch-e' if efficient else 'FlatMatch'}):"
    )
    print(f"  Accuracy: {test_metrics['accuracy']:.4f}")
    print(f"  Precision: {test_metrics['precision']:.4f}")
    print(f"  Recall: {test_metrics['recall']:.4f}")
    print(f"  F1 Score: {test_metrics['f1']:.4f}")

    # Save detailed metrics to CSV
    metrics_df = pd.DataFrame(
        {
            "epoch": list(range(1, len(val_metrics_history) + 1)),
            "val_loss": [m["loss"] for m in val_metrics_history],
            "val_accuracy": [m["accuracy"] for m in val_metrics_history],
            "val_precision": [m["precision"] for m in val_metrics_history],
            "val_recall": [m["recall"] for m in val_metrics_history],
            "val_f1": [m["f1"] for m in val_metrics_history],
        }
    )

    metrics_df.to_csv(
        os.path.join(
            OUTPUT_DIR, f"validation_metrics_{experiment_name}_{num_labeled}.csv"
        ),
        index=False,
    )

    return model, test_metrics, flatmatch.train_metrics


def plot_metrics_comparison(
    standard_results, efficient_results, sample_sizes, metric_name, ylabel, title
):
    plt.figure(figsize=(10, 6))

    # Check if we have results to plot for both methods
    if len(standard_results) != len(sample_sizes) or len(efficient_results) != len(
        sample_sizes
    ):
        print(f"Warning: Results arrays for {metric_name} have mismatched lengths")

        # Create padded arrays with None for missing values
        padded_standard = [None] * len(sample_sizes)
        padded_efficient = [None] * len(sample_sizes)

        # Fill in available values
        for i in range(min(len(standard_results), len(sample_sizes))):
            padded_standard[i] = standard_results[i]

        for i in range(min(len(efficient_results), len(sample_sizes))):
            padded_efficient[i] = efficient_results[i]

        # Use padded arrays for plotting
        standard_results = padded_standard
        efficient_results = padded_efficient

    # Plot available data points (skipping None values)
    std_x = []
    std_y = []
    eff_x = []
    eff_y = []

    for i, size in enumerate(sample_sizes):
        if standard_results[i] is not None:
            std_x.append(size)
            std_y.append(standard_results[i])
        if efficient_results[i] is not None:
            eff_x.append(size)
            eff_y.append(efficient_results[i])

    if std_x:
        plt.plot(std_x, std_y, marker="o", label="FlatMatch", linewidth=2)
    if eff_x:
        plt.plot(
            eff_x, eff_y, marker="s", label="FlatMatch-e", linewidth=2, linestyle="--"
        )

    plt.xlabel("Number of Labeled Samples per Class")
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.xscale("log")
    plt.xticks(sample_sizes, [str(s) for s in sample_sizes])

    plt.savefig(os.path.join(OUTPUT_DIR, f"comparison_{metric_name}.png"))
    plt.close()


# -----------------------------------------------------
# Main Experiment
# -----------------------------------------------------
def main():
    # Initialize tokenizer
    tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

    # Define sampling sizes
    # For the full experiment, use all four sizes
    sample_sizes = [40, 100, 250, 500]

    # Store results for both methods
    standard_results = {"accuracy": [], "precision": [], "recall": [], "f1": []}
    efficient_results = {"accuracy": [], "precision": [], "recall": [], "f1": []}

    # Run experiments with both FlatMatch and FlatMatch-e
    for efficient in [True, False]:
        method_name = "flatmatch_e" if efficient else "flatmatch"
        results_dict = efficient_results if efficient else standard_results

        for num_samples in sample_sizes:
            print(f"\n{'='*80}")
            print(
                f"Running experiment with {num_samples} labeled samples per class using {'FlatMatch-e' if efficient else 'FlatMatch'}"
            )
            print(f"{'='*80}")

            # Create the full dataset for train data
            train_dataset = YahooAnswersDataset(
                "/kaggle/input/yahooanswerssplited/text/train",
                tokenizer,
                max_length=128,
                max_files=1000,
            )

            # Split into labeled and unlabeled datasets
            labeled_dataset, unlabeled_dataset = create_ssl_data_split(
                train_dataset, num_labeled_per_class=num_samples
            )

            # Create dataloaders
            labeled_loader = DataLoader(
                labeled_dataset, batch_size=16, shuffle=True, num_workers=2
            )

            unlabeled_loader = DataLoader(
                unlabeled_dataset,
                batch_size=32,  # Larger batch size for unlabeled data
                shuffle=True,
                num_workers=2,
            )

            # Create validation and test loaders
            val_loader = create_dataloaders(
                "/kaggle/input/yahooanswerssplited/text/val",
                tokenizer,
                max_length=128,
                batch_size=32,
            )

            test_loader = create_dataloaders(
                "/kaggle/input/yahooanswerssplited/text/test",
                tokenizer,
                max_length=128,
                batch_size=32,
            )

            # Train and evaluate
            _, metrics, _ = train_and_evaluate(
                labeled_loader,
                unlabeled_loader,
                val_loader,
                test_loader,
                num_samples,
                epochs=10,
                experiment_name=f"{method_name}",
                efficient=efficient,
            )

            # Store results
            for metric_name in ["accuracy", "precision", "recall", "f1"]:
                results_dict[metric_name].append(metrics[metric_name])

        # Create comparison plots
        for metric_name in ["accuracy", "precision", "recall", "f1"]:
            plot_metrics_comparison(
                standard_results[metric_name],
                efficient_results[metric_name],
                sample_sizes,
                metric_name,
                metric_name.capitalize(),
                f"{metric_name.capitalize()} Comparison: FlatMatch vs FlatMatch-e",
            )

        # Save final results to CSV
        results_df = pd.DataFrame(
            {
                "samples_per_class": sample_sizes + sample_sizes,
                "method": ["FlatMatch"] * len(sample_sizes)
                + ["FlatMatch-e"] * len(sample_sizes),
                "accuracy": standard_results["accuracy"]
                + efficient_results["accuracy"],
                "precision": standard_results["precision"]
                + efficient_results["precision"],
                "recall": standard_results["recall"] + efficient_results["recall"],
                "f1": standard_results["f1"] + efficient_results["f1"],
            }
        )

    results_df.to_csv(os.path.join(OUTPUT_DIR, "final_results.csv"), index=False)
    print("Experiment completed. Results saved to", OUTPUT_DIR)


if __name__ == "__main__":
    main()
```
